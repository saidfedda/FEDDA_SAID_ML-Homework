```python
# =========================================================
# ML & AutoML HOMEWORK
# Nom: [Votre Nom] - PrÃ©nom: [Votre PrÃ©nom]
# =========================================================

# ======================
# INSTALLATION
# ======================
!pip install -q imbalanced-learn tpot

# ======================
# IMPORT
# ======================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from imblearn.over_sampling import SMOTE
import warnings
warnings.filterwarnings('ignore')

# ======================
# I) PROJET 1 : SEGMENTATION CLIENTÃˆLE
# ======================
print("="*60)
print("PROJET 1 : SEGMENTATION CLIENTÃˆLE")
print("="*60)

# 1. Charger et explorer le dataset
df = pd.read_csv("/content/Mall_Customers.csv")
df.columns = ["CustomerID", "Genre", "Age", "Annual_Income", "Spending_Score"]

print("\nAperÃ§u des donnÃ©es:")
print(df.head())
print(f"\nNombre de clients: {len(df)}")

# 2. Analyser les caractÃ©ristiques des clients
print("\nStatistiques descriptives:")
print(df.describe())

# Visualisations
fig, axes = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Analyse des CaractÃ©ristiques des Clients', fontsize=14)

axes[0, 0].hist(df["Age"], bins=20, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Distribution Ã‚ge')
axes[0, 0].set_xlabel('Ã‚ge')

axes[0, 1].hist(df["Annual_Income"], bins=20, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Distribution Revenu Annuel')
axes[0, 1].set_xlabel('Revenu (k$)')

axes[1, 0].hist(df["Spending_Score"], bins=20, color='salmon', edgecolor='black')
axes[1, 0].set_title('Distribution Score de DÃ©pense')
axes[1, 0].set_xlabel('Score')

df["Genre"].value_counts().plot(kind='pie', ax=axes[1, 1], autopct='%1.1f%%')
axes[1, 1].set_title('RÃ©partition Genre')
axes[1, 1].set_ylabel('')

plt.tight_layout()
plt.show()

# Encoder le genre
df["Genre"] = df["Genre"].map({"Male": 0, "Female": 1})

# 3. Appliquer les mÃ©thodes de clustering
X_clust = df[["Age", "Annual_Income", "Spending_Score"]]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clust)

# K-Means
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df["KMeans_Cluster"] = kmeans.fit_predict(X_scaled)

# AgglomÃ©ratif
agg = AgglomerativeClustering(n_clusters=5)
df["Agglomerative_Cluster"] = agg.fit_predict(X_scaled)

# DBSCAN
dbscan = DBSCAN(eps=1.2, min_samples=5)
df["DBSCAN_Cluster"] = dbscan.fit_predict(X_scaled)

# 4. Visualiser les segments (scatter plots)
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
fig.suptitle('Comparaison des MÃ©thodes de Clustering', fontsize=14)

# K-Means
for cluster in df["KMeans_Cluster"].unique():
    subset = df[df["KMeans_Cluster"] == cluster]
    axes[0].scatter(subset["Annual_Income"], subset["Spending_Score"],
                    label=f"C{cluster}", alpha=0.6)
axes[0].set_title('K-Means')
axes[0].set_xlabel('Revenu Annuel')
axes[0].set_ylabel('Score de DÃ©pense')
axes[0].legend()

# AgglomÃ©ratif
for cluster in df["Agglomerative_Cluster"].unique():
    subset = df[df["Agglomerative_Cluster"] == cluster]
    axes[1].scatter(subset["Annual_Income"], subset["Spending_Score"],
                    label=f"C{cluster}", alpha=0.6)
axes[1].set_title('AgglomÃ©ratif')
axes[1].set_xlabel('Revenu Annuel')
axes[1].legend()

# DBSCAN
for cluster in df["DBSCAN_Cluster"].unique():
    subset = df[df["DBSCAN_Cluster"] == cluster]
    label = "Bruit" if cluster == -1 else f"C{cluster}"
    axes[2].scatter(subset["Annual_Income"], subset["Spending_Score"],
                    label=label, alpha=0.6)
axes[2].set_title('DBSCAN')
axes[2].set_xlabel('Revenu Annuel')
axes[2].legend()

plt.tight_layout()
plt.show()

# Pair plot pour K-Means
sns.pairplot(df[["Age", "Annual_Income", "Spending_Score", "KMeans_Cluster"]],
             hue="KMeans_Cluster", palette="Set2")
plt.suptitle('Pair Plot - K-Means', y=1.02)
plt.show()

# 5. InterprÃ©ter les segments et recommandations marketing
print("\nStatistiques par segment (K-Means):")
cluster_stats = df.groupby('KMeans_Cluster')[['Age', 'Annual_Income', 'Spending_Score']].mean()
print(cluster_stats)

print("\nðŸ’¡ RECOMMANDATIONS MARKETING:")
for cluster in range(5):
    stats = cluster_stats.loc[cluster]
    print(f"\nCluster {cluster}: Ã‚ge={stats['Age']:.0f}, Revenu={stats['Annual_Income']:.0f}k$, Score={stats['Spending_Score']:.0f}")

    if stats['Spending_Score'] > 60 and stats['Annual_Income'] > 60:
        print("   â†’ Clients Premium - Produits haut de gamme")
    elif stats['Spending_Score'] > 60 and stats['Annual_Income'] < 40:
        print("   â†’ DÃ©pensiers modestes - Promotions et fidÃ©litÃ©")
    elif stats['Spending_Score'] < 40:
        print("   â†’ Clients Ã©conomes - Offres accessibles")
    else:
        print("   â†’ Segment moyen - Marketing Ã©quilibrÃ©")

# ======================
# II) PROJET 2 : DÃ‰TECTION DE FRAUDE
# ======================
print("\n" + "="*60)
print("PROJET 2 : DÃ‰TECTION DE FRAUDE")
print("="*60)

# 1. Charger et explorer le dataset
fraud = pd.read_csv("/content/creditcard.csv")
fraud = fraud.sample(50000, random_state=42)

print("\nAperÃ§u des donnÃ©es:")
print(fraud.head())
print(f"\nNombre de transactions: {len(fraud)}")

# 2. Analyser la distribution des classes
print("\nDistribution des classes:")
print(fraud['Class'].value_counts())
print(f"Pourcentage de fraudes: {(fraud['Class'].sum()/len(fraud))*100:.2f}%")

# Visualisation
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

fraud['Class'].value_counts().plot(kind='bar', ax=axes[0], color=['green', 'red'])
axes[0].set_title('Distribution Classes (Avant SMOTE)')
axes[0].set_xticklabels(['Normale', 'Fraude'], rotation=0)

axes[1].hist([fraud[fraud['Class']==0]['Amount'], fraud[fraud['Class']==1]['Amount']],
             bins=50, label=['Normale', 'Fraude'], color=['green', 'red'], alpha=0.7)
axes[1].set_title('Distribution Montants')
axes[1].set_xlabel('Montant')
axes[1].legend()
axes[1].set_yscale('log')

plt.tight_layout()
plt.show()

# 3. PrÃ©parer les donnÃ©es
X = fraud.drop("Class", axis=1)
y = fraud["Class"]

# Normalisation
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Gestion du dÃ©sÃ©quilibre avec SMOTE
smote = SMOTE(random_state=42)
X_res, y_res = smote.fit_resample(X_scaled, y)
print(f"\nAprÃ¨s SMOTE: Classe 0={(y_res==0).sum()}, Classe 1={(y_res==1).sum()}")

# Division train/test
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42
)

# 4. Construire les modÃ¨les de classification

# Logistic Regression
lr = LogisticRegression(max_iter=1000)
lr.fit(X_train, y_train)
y_pred_lr = lr.predict(X_test)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

# 5. Ã‰valuer les modÃ¨les
print("\n--- LOGISTIC REGRESSION ---")
print(classification_report(y_test, y_pred_lr))

print("\n--- RANDOM FOREST ---")
print(classification_report(y_test, y_pred_rf))

# Matrices de confusion
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

cm_lr = confusion_matrix(y_test, y_pred_lr)
sns.heatmap(cm_lr, annot=True, fmt='d', cmap='Blues', ax=axes[0])
axes[0].set_title('Logistic Regression')

cm_rf = confusion_matrix(y_test, y_pred_rf)
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Greens', ax=axes[1])
axes[1].set_title('Random Forest')

plt.tight_layout()
plt.show()

# 6. InterprÃ©ter et amÃ©liorer
print("\nðŸ’¡ INTERPRÃ‰TATION:")
print("- Les deux modÃ¨les ont de bonnes performances aprÃ¨s SMOTE")
print("- Random Forest Ã©quilibre mieux prÃ©cision et rappel")

print("\nðŸ”§ AMÃ‰LIORATIONS POSSIBLES:")
print("- Optimiser les hyperparamÃ¨tres (Grid Search)")
print("- Tester d'autres algorithmes (XGBoost, SVM)")
print("- Feature engineering")
print("- Utiliser AutoML pour automatiser l'optimisation")

# ======================
# III) Ã‰TAT DE L'ART AutoML
# ======================
print("\n" + "="*60)
print("III) Ã‰TAT DE L'ART SUR AutoML")
print("="*60)

print("""
ðŸ“š QU'EST-CE QUE L'AutoML ?

AutoML (Automated Machine Learning) automatise les Ã©tapes du ML :
â€¢ PrÃ©traitement des donnÃ©es
â€¢ SÃ©lection des features
â€¢ Choix du modÃ¨le
â€¢ Optimisation des hyperparamÃ¨tres

ðŸŽ¯ INTÃ‰RÃŠT :
â€¢ Gain de temps
â€¢ Accessible aux non-experts
â€¢ Performances optimales

ðŸ› ï¸ TPOT (Tree-based Pipeline Optimization Tool)

FONCTIONNEMENT:
â€¢ Utilise des algorithmes gÃ©nÃ©tiques
â€¢ Explore diffÃ©rents pipelines automatiquement
â€¢ Teste plusieurs modÃ¨les et hyperparamÃ¨tres

AVANTAGES:
âœ… Open source
âœ… IntÃ©gration scikit-learn
âœ… Export du code Python

LIMITES:
âŒ Temps de calcul Ã©levÃ©
âŒ Consomme beaucoup de ressources

ðŸ› ï¸ AUTO-SKLEARN

FONCTIONNEMENT:
â€¢ Optimisation bayÃ©sienne
â€¢ Utilise mÃ©tadonnÃ©es d'expÃ©riences passÃ©es
â€¢ Ensemble automatique de modÃ¨les

AVANTAGES:
âœ… Plus rapide que TPOT
âœ… Gestion automatique des donnÃ©es manquantes

LIMITES:
âŒ Installation complexe
âŒ Moins flexible
""")

# ======================
# IV) APPLICATION AutoML
# ======================
print("\n" + "="*60)
print("IV) APPLICATION AutoML AVEC TPOT")
print("="*60)

from tpot import TPOTClassifier

# 1. DonnÃ©es dÃ©jÃ  prÃ©parÃ©es (mÃªme Ã©tapes que manuel)
print("\nDonnÃ©es prÃªtes pour AutoML")
print(f"X_train: {X_train.shape}, X_test: {X_test.shape}")

# 2. Lancer TPOT
print("\nðŸ¤– Lancement TPOT ...")

tpot = TPOTClassifier(
    generations=3,
    population_size=2,
    max_time_mins=10,
    random_state=42,
    n_jobs=6,

)

tpot.fit(X_train[:30000], y_train[:30000])
y_pred_tpot = tpot.predict(X_test)

# 3. Comparer les performances
print("\n--- TPOT AutoML ---")
print(classification_report(y_test, y_pred_tpot))

# Comparaison
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

comparison = pd.DataFrame({
    'ModÃ¨le': ['Logistic Regression', 'Random Forest', 'TPOT AutoML'],
    'PrÃ©cision': [accuracy_score(y_test, y_pred_lr),
                  accuracy_score(y_test, y_pred_rf),
                  accuracy_score(y_test, y_pred_tpot)],
    'Rappel': [recall_score(y_test, y_pred_lr),
               recall_score(y_test, y_pred_rf),
               recall_score(y_test, y_pred_tpot)],
    'F1-Score': [f1_score(y_test, y_pred_lr),
                 f1_score(y_test, y_pred_rf),
                 f1_score(y_test, y_pred_tpot)]
})

print("\nðŸ“Š COMPARAISON:")
print(comparison.to_string(index=False))

comparison.set_index('ModÃ¨le').plot(kind='bar', figsize=(10, 5), rot=0)
plt.title('Comparaison ModÃ¨les Manuels vs AutoML')
plt.ylabel('Score')
plt.ylim([0, 1.1])
plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# 4. DÃ©crire le pipeline final
print("\nðŸ” PIPELINE FINAL TROUVÃ‰ PAR TPOT:")
print(tpot.fitted_pipeline_)

print("\nDESCRIPTION:")
print("TPOT a automatiquement trouvÃ© le meilleur pipeline incluant:")
print("- PrÃ©processeurs optimaux")
print("- Meilleur modÃ¨le")
print("- HyperparamÃ¨tres optimisÃ©s")

# Export
print("Best pipeline found by TPOT:")
print(tpot.fitted_pipeline_)

# ======================
# CONCLUSION
# ======================
print("\n" + "="*60)
print("âœ¨ HOMEWORK TERMINÃ‰ âœ¨")
print("="*60)

print("""
âœ… Projet 1: Segmentation avec 3 mÃ©thodes de clustering
âœ… Projet 2: DÃ©tection de fraude avec 2 modÃ¨les
âœ… AutoML: Comparaison approche manuelle vs automatique
âœ… Toutes les visualisations et analyses complÃ¨tes
""")
```
